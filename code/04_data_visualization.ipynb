```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from jupyterthemes import jtplot
jtplot.style()
```

```python
df=sns.load_dataset('flights')
```

```python
df.head()
```

```python
f=plt.figure(figsize=(15,8))
plt.bar(df['year'],df['passengers']);
```

```python
f=plt.figure(figsize=(15,8))
sns.barplot(data=df,x='year',y='passengers');
```

```python
def get_date(data):
    data=data.to_dict()
    month = datetime.strptime(data['month'],'%b').month
    return f"{data['year']}-{month}"
```

```python
df['date']=df[['year','month']].apply(get_date,axis=1)
```

```python
df
```

```python
plt.figure(figsize=(35,12))
plt.plot(df['date'],df['passengers'])
plt.xlim(0,144)
plt.xticks(rotation=60);
```

```python
plt.figure(figsize=(35,12))
sns.lineplot(data=df,x='date',y='passengers')
plt.xlim(0,144)
plt.xticks(rotation=60);
```

```python
plt.figure(figsize=(15,8))
plt.boxplot(df['passengers']);
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df,x='passengers');
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=df,y='passengers',x='year');
plt.xticks(rotation=60);
```
